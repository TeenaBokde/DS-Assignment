# Standardization of Numerical Variables

Standardization (Z-Score Normalization):

Standardisation, also known as Z-score normalisation, is a crucial concept in data preprocessing, particularly in machine learning and statistics.
This technique rescales features to a mean of 0 and a standard deviation of 1, thereby preventing features with large ranges from dominating the model.

This problem is solved by standardisation, which gives each feature a mean of 0 and a standard deviation of 1.

This is gain by subtracting the mean and dividing by the standard deviation for each feature.
The formula for standardization is:

z = (x - μ) / σ

where:

z = standardized value,
x = original value,
μ = mean of the feature,
σ = standard deviation of the feature

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
numerical_cols = ['Volume', 'Avg Price', 'Total Sales Value', 'Discount Rate (%)', 'Discount Amount', 'Net Sales Value']
```

```python
print(numerical_cols)
```

```python
df[numerical_cols] = scaler.fit_transform(df[numerical_cols])
```

```python
def standardize(x):
    return (x - x.mean()) / x.std()

df[numerical_cols] = df[numerical_cols].apply(standardize)
```

# Conversion of Categorical Data into Dummy Variables

Discuss the need for converting categorical data into dummy variables (one-hot encoding).

When working with machine learning models, it's essential to preprocess categorical data into a format that can be understood by the algorithm.
One common technique used for this purpose is one-hot encoding, also known as dummy variables.

Why do we need to convert categorical data?

Categorical data, such as strings or categorical values, cannot be directly fed into machine learning models. This is because most algorithms are designed to work with numerical data. Moreover, categorical data doesn't have a natural ordering or scale, making it difficult for models to understand the relationships between different categories.

What are dummy variables?

Dummy variables, or one-hot encoding, is a technique that converts categorical data into a numerical representation. Each category is represented as a binary vector, where all elements are 0 except for one element, which is 1. This creates a new feature for each category, allowing the model to treat each category as a separate feature.

Benefits of one-hot encoding:

Model interpretability: One-hot encoding allows us to visualize and understand the relationships between different categories.

Improved model performance: By converting categorical data into numerical features, we can improve the accuracy and performance of machine learning models.

Handling categorical data with multiple levels: One-hot encoding can handle categorical data with multiple levels, making it a versatile technique for preprocessing categorical data.

```python
# Create dummy variables for categorical columns
data = pd.get_dummies(data, columns=['Date', 'Day', 'SKU', 'City', 'BU', 'Brand', 'Model'])
data
```

Data Preprocessing: The Unsung Hero of Data Analysis and Machine Learning


Data preprocessing is a crucial step in the data analysis and machine learning pipeline. It involves transforming and preparing the data into a format that is suitable for modeling. Two essential preprocessing steps are standardization and one-hot encoding.

Standardization


Standardization, also known as normalization, is the process of scaling numerical features to a common range, usually between 0 and 1. This is necessary because many machine learning algorithms are sensitive to the scale of the features.

For example, if one feature has a large range of values, it may dominate the model, leading to poor performance.

Standardization helps to:


- Prevent feature dominance
- Improve model performance
- Enhance interpretability

One-Hot Encoding


One-hot encoding is a technique used to convert categorical features into a numerical representation that can be processed by machine learning algorithms. It involves creating a binary vector for each category, where all elements are 0 except for one element, which is 1.

One-hot encoding helps to:


- Convert categorical features into a numerical format
- Avoid ordinality assumptions (e.g., assuming that 'A' is greater than 'B')
- Improve model performance


In Python, you can use the get_dummies function from pandas to one-hot encode your data

- Improves model performance by reducing the effect of noise and outliers
- Enhances interpretability by making it easier to understand the relationships between features
